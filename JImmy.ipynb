```python
import  numpy as np
import pandas as pd 
import matplotlib .pyplot as plt

dataframe = pd.read_csv("space_Corrected.csv")
df=dataframe.dropna()

df

ssa = df.loc[df['Status Rocket'] == 'StatusActive']
gbs = ssa.groupby('Company Name')['Status Rocket'].count().reset_index()

print(gbs)
fig = plt.subplots(figsize=(16,8))
plt.title('Hey Jim')
plt.xlabel('Company Name')
plt.ylabel('Active Rockets')
plt.xticks(rotation='vertical')
label = gbs['Company Name']
value = gbs['Status Rocket']
plt.bar(label, value,width=0.4, color=('red','blue','green','pink','yellow'))
```
